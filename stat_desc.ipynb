```python
# packages
import pandas as pd
import plotly.express as px
```

# Data management

```python
# Chargement des données
df = pd.read_csv('data_inachevee.csv')
```

```python
# Calcul du score humain moyen
df["score_humain"] = df[["Garance", "Matthias", "Yannis"]].mean(axis=1) / 10
```

```python
# Nombre de charactères par contribution
df["nb_char"] = df["contribution"].apply(len)
```

# Statistiques descriptives
1. ***Distribution du nombre de charactères par contribution***

```python
# Barplot du nombre de charactères par contribution sur les 200 premières contributions
fig = px.histogram(
    df, x="nb_char", nbins=max(df["nb_char"])//100,
    color_discrete_sequence=["#2a2781"]
)
fig.update_layout(
    title={"text": "<b>Distribution du nombre de charactères par contribution</b>"}, 
    xaxis_title="Nombre de charactères", 
    yaxis_title="Nombre de contributions", 
    width=800, height=450
)
fig.update_traces(
    hovertemplate =
        "Nombre de charactères = %{x}<br>" \
        "Nombre de contributions = %{y}<extra></extra>"
)
fig.show()
```

3. ***Distribution du nombre d'hallucinations en fonction du nombre de charactères par contribution***

```python
fig = px.histogram(
    df, x="nb_char", y="Hallucinations", nbins=max(df["nb_char"])//100,
    color_discrete_sequence=["#2a2781"]
)
fig.update_layout(
    title={"text": "<b>Distribution du nombre d'hallucinations en fonction du nombre de <br>charactères par contribution</b>"}, 
    xaxis_title="Nombre de charactères", 
    yaxis_title="Nombre d'hallucinations", 
    width=800, height=450
)
fig.update_traces(
    hovertemplate =
        "Nombre de charactères = %{x}<br>" \
        "Nombre d'hallucinations = %{y}<extra></extra>"
)
fig.show()
```

4. ***Distribution du nombre d'idées non indépendantes en fonction du nombre de charactères par contribution***

```python
fig = px.histogram(
    df, x="nb_char", y="Idées_non_ind", nbins=max(df["nb_char"])//100,
    color_discrete_sequence=["#2a2781"]
)
fig.update_layout(
    title={"text": "<b>Distribution du nombre d'idées non indépendantes en fonction du <br>nombre de charactères par contribution</b>"}, 
    xaxis_title="Nombre de charactères", 
    yaxis_title="Nombre d'idées non ind", 
    width=800, height=450
)
fig.update_traces(
    hovertemplate =
        "Nombre de charactères = %{x}<br>" \
        "Nombre d'idées non ind = %{y}<extra></extra>"
)
fig.show()
```

5. ***Score humain moyen vs métrique QualIT***

```python
fig = px.scatter(
    df, x="C", y="score_humain", 
    color_discrete_sequence=["#2a2781"]
)
# Ajout de la ligne x = y pour référence
fig.add_shape(
    type="line", x0=0, x1=1, y0=0, y1=1, xref="x", yref="y",
    line=dict(color="#d41010")
)
fig.update_layout(
    title={"text": "<b>Comparaison du score humain avec la métrique QualIT</b>"},
    xaxis_title="Métrique QualIT",
    yaxis_title="Score humain",
    width=700, height=500
)
fig.update_traces(
    hovertemplate =
        "Métrique QualIT = %{x}<br>" \
        "Score humain = %{y}<extra></extra>"
)
fig.show()
```
